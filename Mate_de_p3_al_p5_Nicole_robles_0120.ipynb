#### Problema 3:
Resta y resultado
* Un número al que se le resta 7 da como resultado 18.
* Planteamiento: Sea $x$ el número desconocido. Si le quitamos 7, obtenemos 18.
* Ecuación $x - 7 = 18
* Despeje: 7 está restando; pasa al otro lado de la igualdad sumando.$x = 18 + 7$
* Solución: $x = 25$

```python
# Problema 3: x - 7 = 18
resultado = 18
sustraendo = 7

# Despejamos x sumando el sustraendo al resultado
x = resultado + sustraendo

print(f"El número del Problema 3 es: {x}")
```

#### Problema 4:
Triple de un número más una constante
* El triple de un número, más 4 es igual a 31
* Planteamiento: "El triple de un número" se expresa como $3x$. "Más 4" se añade a esa expresión.
* Ecuación: $3x + 4 = 31$
* Despeje: * Primero, movemos el 4 (que suma) restando: $3x = 31 - 4 \Rightarrow 3x = 27$Luego, el 3 (que multiplica) pasa dividiendo: $x = \frac{27}{3}$Solución: $x = 9$

```python
# Problema 4: 3x + 4 = 31
total = 31
constante = 4
coeficiente = 3

# Paso 1: Restar la constante. Paso 2: Dividir por el coeficiente
x = (total - constante) / coeficiente

print(f"El número del Problema 4 es: {int(x)}")
```

#### Problema 5:
Dimensiones de un terreno (Perímetro)Un terreno tiene un largo que mide 5 metros más que el ancho, si el perímetro es de 50 metros.
* Planteamiento:Ancho: $x$Largo: $x + 5$El perímetro de un rectángulo es la suma de todos sus lados: $P = 2(largo) + 2(ancho)$.
* Ecuación:$2(x + 5) + 2(x) = 50$
* Simplificando: $2x + 10 + 2x = 50 \Rightarrow 4x + 10 = 50$
* Despeje:Restamos 10: $4x = 50 - 10 \Rightarrow 4x = 40$Dividimos por 4: $x = \frac{40}{4}$
* Solución:Ancho ($x$): 10 metros.Largo ($x + 5$): 15 metros.

```python
# Problema 5: 4x + 10 = 50
perimetro = 50
diferencia_largo = 5
# La ecuación simplificada es 4x + (2 * diferencia_largo) = perimetro

# Despejamos x (ancho)
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Dimensiones: Ancho = {int(ancho)}m, Largo = {int(largo)}m")
```
