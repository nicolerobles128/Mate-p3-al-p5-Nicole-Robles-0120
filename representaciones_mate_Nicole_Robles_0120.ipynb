#### Problema 1
 La suma de un número más nueve es igual a 21
 * Paso 1: Explicación paso a paso
 * Planteamiento del problema:Identificamos la incógnita: No conocemos "el número", así que lo representamos con la variable $x$.
 * Traducción al lenguaje algebraico: "La suma de un número ($x$) más el número nueve ($+ 9$)" es igual ($=$) a $21$.La ecuación:$$x + 9 = 21$$Despeje de la variable:Para dejar a la $x$ sola, el $9$ que está sumando pasa al otro lado del signo igual con la operación contraria (resta).$x = 21 - 9$
 * Solución:$x = 12$
 * Comprobación: $12 + 9 = 21$. El número es 12.

```python
# Problema 1: x + 9 = 21
resultado_total = 21
numero_conocido = 9

# Despejamos x: x = resultado_total - numero_conocido
x = resultado_total - numero_conocido

print(f"El número buscado en el Problema 1 es: {x}")
```

#### Problema 2
 Edades de Pedro y Ana
 * Paso 1: Explicación paso a pasoPlanteamiento del problema:Definimos variables basadas en la relación:Edad de Ana: $x$Edad de Pedro: $2x$ (ya que tiene el doble que Ana).
 * Sabemos que la suma de ambas edades es $36$.La ecuación:$$x + 2x = 36$$Simplificamos sumando términos semejantes: $3x = 36$.
 * Despeje de la variable:El $3$ que está multiplicando a la $x$ pasa al otro lado dividiendo.$x = \frac{36}{3}$
 * Solución:$x = 12$ (Edad de Ana).
 * Pedro tiene el doble: $2 \times 12 = 24$.
 * Comprobación: $12 + 24 = 36$.
 * Ana tiene 12 años y Pedro tiene 24 años.

```python
# Problema 2: x + 2x = 36  => 3x = 36
suma_edades = 36
proporcion_total = 1 + 2  # Ana (1) + Pedro (2)
resultado_total=int(input("Ingrese el resultado total:"))
numero_conocido=int(input("Ingrese el numero conocido:"))
# Despejamos x: x = suma_edades / proporcion_total
# Calculamos la unidad x (edad de Ana)
edad_ana = suma_edades / proporcion_total
edad_pedro = edad_ana * 2

print(f"Ana tiene {int(edad_ana)} años")
print(f"Pedro tiene {int(edad_pedro)} años.")
```
